```python
import os
from pathlib import Path
import pandas as pd

folder_path = '2016_original/all_csv'

airports_of_interest = ['ATL', 'CLT', 'DEN', 'DFW', 'EWR', 'IAH', 'JFK', 'LAS', 'LAX', 'MCO', 'MIA', 'ORD', 'PHX', 'SEA', 'SFO']

folder = Path(folder_path)
result_df = pd.DataFrame()
for file_path in folder.glob('*.csv'):
    
    df = pd.read_csv(file_path)

    
    filtered_df = df[df['Origin'].isin(airports_of_interest) & df['Dest'].isin(airports_of_interest)]

    selected_columns = ['FlightDate', 'Quarter', 'Year', 'Month', 'DayofMonth', 'DepTime', 'DepDel15',
                         'CRSDepTime', 'DepDelayMinutes', 'Origin', 'Dest', 'ArrTime', 'CRSArrTime', 'ArrDel15',
                         'ArrDelayMinutes']

    
    result_df = pd.concat([result_df, filtered_df[selected_columns]])



print(result_df.head())
```
